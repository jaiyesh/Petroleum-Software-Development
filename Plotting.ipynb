```python
import numpy as np
import matplotlib.pyplot as plt
```

### Exercise: Generate Relative Permeability plots in python

```python
sw = np.linspace(0,1,100)
```

```python
sw
```

```python
krw = sw**3
```

```python
krw
```

```python
kro = (1-sw)**2
```

```python
kro
```

```python
#step1: Krw vs Sw plot
plt.plot(sw,krw)
```

```python
#step2 : Kro vs Sw plot
plt.plot(sw,kro)
```

```python
#Step 3: both lines on the same figure

plt.plot(sw,krw)
plt.plot(sw,kro)
```

```python
#step 4: give x axis label

plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("Sw(Fraction)")
```

```python
#step 5: give y axis label

plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("Sw(Fraction)")
plt.ylabel("Relative Permeability(Fraction)")
```

```python
#step 6: LEGEND

plt.plot(sw,krw, label = "Water Relative Permeability")
plt.plot(sw,kro, label = "Oil Relative Permeability")

plt.xlabel("Sw(Fraction)")
plt.ylabel("Relative Permeability(Fraction)")
plt.legend()
```

```python
#step 7: grid

plt.plot(sw,krw, label = "Water Relative Permeability")
plt.plot(sw,kro, label = "Oil Relative Permeability")

plt.xlabel("Sw(Fraction)")
plt.ylabel("Relative Permeability(Fraction)")
plt.legend()

plt.grid()
```

```python
#step 8: Change the size of the plot

plt.figure(figsize=(5,3))

plt.plot(sw,krw, label = "Water Relative Permeability")
plt.plot(sw,kro, label = "Oil Relative Permeability")

plt.xlabel("Sw(Fraction)")
plt.ylabel("Relative Permeability(Fraction)")
plt.legend()

plt.grid()
```

```python
#step 9: Change the line-color

plt.figure(figsize=(5,3))

plt.plot(sw,krw, label = "Water Relative Permeability",color="darkblue")
plt.plot(sw,kro, label = "Oil Relative Permeability",color = "brown")

plt.xlabel("Sw(Fraction)")
plt.ylabel("Relative Permeability(Fraction)")
plt.legend()

plt.grid()
```

```python
#step 10: Change the line-thickness

plt.figure(figsize=(5,3))

plt.plot(sw,krw, label = "Water Relative Permeability",color="darkblue",lw=2)
plt.plot(sw,kro, label = "Oil Relative Permeability",color = "brown",lw=5)

plt.xlabel("Sw(Fraction)")
plt.ylabel("Relative Permeability(Fraction)")
plt.legend()

plt.grid()
```

```python
#step 11: Adding title to the plot

plt.figure(figsize=(7,5))

plt.plot(sw,krw, label = "Water Relative Permeability",color="darkblue",lw=2)
plt.plot(sw,kro, label = "Oil Relative Permeability",color = "brown",lw=5)

plt.xlabel("Sw(Fraction)")
plt.ylabel("Relative Permeability(Fraction)")
plt.legend()
plt.title("Relative Permeability Plot")
plt.grid()
```

```python
#step 12: styles
plt.style.available
```

```python

plt.style.use('classic')

plt.figure(figsize=(7,5))

plt.plot(sw,krw, label = "Water Relative Permeability",color="darkblue",lw=2)
plt.plot(sw,kro, label = "Oil Relative Permeability",color = "brown",lw=5)

plt.xlabel("Sw(Fraction)")
plt.ylabel("Relative Permeability(Fraction)")
plt.legend()
plt.title("Relative Permeability Plot")
plt.grid(True)
```

## Subplots: Multiple plots next to each other

```python
#step 1: make axis and figures for number of rows and columns
fig , axs = plt.subplots(nrows=2, ncols=2);
## step2: figure size
fig.set_figheight(6) ; fig.set_figwidth(10)

## step3: x data
x_data = np.linspace(1,100,1000)
## step4 ploting
axs[0,0].plot(x_data, x_data**1, label='Power = 1')
axs[0,0].set_title('Power = 1')
axs[0,0].grid()

axs[0,1].plot(x_data, x_data**2, label='Power = 2', color='red')
axs[0,1].set_title('Power = 2')

axs[1,0].plot(x_data, x_data**3, label='Power = 3', color='brown')
axs[1,0].set_title('Power = 3')

# axs[1,1].plot(x_data, x_data**4, label='Power = 4', color='orange')
axs[1,1].hist(porosity)
axs[1,1].set_title('Porosity Histogrom')

fig.tight_layout()
```
